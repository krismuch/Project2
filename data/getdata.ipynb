```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Import weather csv files, first county then daily weather
file = "county.csv"
county_df = pd.read_csv(file)
county_df.head()
```

```python
file2 = "dailyweather.csv"
stats_df = pd.read_csv(file2)
stats_df.head()
```

# Overall by Month

```python
# Derive MONTH and YEAR variables from DATUM
stats_df["Test"] = stats_df.DATUM.str[5:]
stats_df["YEAR"] = stats_df.DATUM.str[:4]
stats_df["MONTH"] = stats_df.Test.str[:2]
stats_df.head()
```

```python
# Group by MONTH and FIPS (county) to calculate monthly average and cumulative rain total
bycounty_bymonth = stats_df.groupby(['FIPS','MONTH'])
weather_overall_df1 = bycounty_bymonth['TMAX','TMIN'].mean().reset_index()
weather_overall_df2 = bycounty_bymonth['PRCP'].sum().reset_index()
weather_overall_df3 = pd.merge(weather_overall_df1, weather_overall_df2,  how='inner', on=['FIPS','MONTH'])
# Represent OVERALL stats with the YEAR 9999
weather_overall_df3["YEAR"]="9999"
weather_overall_df3.head()
```

# By Year and Month

```python
bycounty_byyear_bymonth = stats_df.groupby(['FIPS','YEAR','MONTH'])
bycounty_byyear_bymonth_avg = bycounty_byyear_bymonth['TMAX','TMIN'].mean().reset_index()
bycounty_byyear_bymonth_sum = bycounty_byyear_bymonth['PRCP'].sum().reset_index()
bycounty_byyear_bymonth_stats = pd.merge(bycounty_byyear_bymonth_avg, bycounty_byyear_bymonth_sum,  how='inner', on=['FIPS','YEAR','MONTH'])
bycounty_byyear_bymonth_stats.head()
```

```python
monthlyweather=pd.concat([weather_overall_df3, bycounty_byyear_bymonth_stats], axis=0, sort=False)
monthlyweather['W_KEY']=monthlyweather.index
monthlyweather.head(100000)
```

```python
monthlyweather.to_csv(r'Data\monthlyweather.csv', index = False)
```
